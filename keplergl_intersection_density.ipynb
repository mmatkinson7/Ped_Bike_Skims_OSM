```python
import pandas as pd 
import geopandas as gpd
import keplergl
```

```python
## load shapefile

url = "J:\Shared drives\TMD_TSA\Data\GIS Data\TAZ\candidate_CTPS_TAZ_STATEWIDE_2019_wgs84.shp"
gdf = gpd.read_file(url)

gdf.head()
```

```python
#load intersection density tables
purl = "J:\Shared drives\TMD_TSA\Data\\Non_Motorized_Networks\Intersections_per_TAZ\ped_intersection_density.csv"
#burl = "J:\Shared drives\TMD_TSA\Data\\Non_Motorized_Networks\Intersections_per_TAZ\\bike_intersection_density.csv"

ped_tab = pd.read_table(purl, sep= ',')
#bike_tab = pd.read_table(burl, sep= ',')

ped_tab = ped_tab.fillna(0)
#bike_tab = bike_tab.fillna(0)
```

```python
#add employment, population, and Land Use type
lurl = "J:\Shared drives\TMD_TSA\Data\\Non_Motorized_Networks\\taz_ru_25.csv"

lu_tab = pd.read_table(lurl, sep=',')
lu_tab = lu_tab.fillna(0)
```

```python
#join bike and ped intersection density tables
ped_gdf = pd.merge(gdf, ped_tab, on='id')
ped_gdf = pd.merge(ped_gdf, lu_tab, left_on = 'id', right_on = 'TAZ_ID')
#bike_gdf = pd.merge(gdf, bike_tab, on='id')
#bike_gdf = pd.merge(bike_gdf, lu_tab, left_on = 'id', right_on = 'TAZ_ID')
ped_gdf.head()
```

```python
#rounding
#bike_gdf = bike_gdf.round({'intersection_density':2, 'Area_SqMi':2})
ped_gdf = ped_gdf.round({'intersection_density':2, 'Area_SqMi':2})
```

```python
#wl_bike = keplergl.KeplerGl(height = 800)
#wl_bike.add_data(data=bike_gdf, name= 'Bike Intersection Density')
wl_ped = keplergl.KeplerGl(height = 800)
wl_ped.add_data(data=ped_gdf, name= 'Pedestrian Intersection Density')
```
